```python
import sqlite3 
import pandas as pd

DB_NAME = "cities.db"

conn = sqlite3.connect(DB_NAME)
```

```python
df = pd.read_sql_query("SELECT * FROM cities", conn)
df
```

```python
df['white_pct'] = df['white'] / df['population'] * 100
df['black_pct'] = df['black'] / df['population'] * 100
df['asian_pct'] = df['asian'] / df['population'] * 100
df['latin_pct'] = df['latin'] / df['population'] * 100
df['low_food_access_pct'] = df['low_food_access'] / df['population'] * 100

df[['city','white_pct','black_pct','asian_pct','latin_pct','low_food_access_pct']]
```
